```python
net = MSMFMultiCAN
experiment_num = 15
file_path = 'tmp.csv'

data = load_experimental_data(ExperimentType.OPTIMIZATION, net, experiment_num)

pos_error_mean_median = np.array(data['pos_error_mean_median'])

min_ids = np.argwhere(pos_error_mean_median == np.min(pos_error_mean_median))

print(np.min(pos_error_mean_median))
print(min_ids)

# for key, value in data.items():
#     print(f'{key}: {value[min_id]}')

min_ids = min_ids.flatten()

# writing to csv file
with open(file_path, 'w') as csvfile:
    # creating a csv writer object
    csvwriter = csv.writer(csvfile)

    # writing the headers if file is newly created
    csvwriter.writerow(data.keys())

    # writing the data rows
    for i in min_ids:
        row = []
        for v in data.values():
            if len(v) > i:
                row.append(v[i])
        csvwriter.writerow(row)

```
